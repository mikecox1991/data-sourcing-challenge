### Import Required Libraries and Set Up Environment Variables

```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()

nyt_api_key = os.getenv("NYT_API_KEY")
tmdb_api_key = os.getenv("TMDB_API_KEY")
```

### Access the New York Times API

```python
# Set the base URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL
query_url = f"{url}api-key={nyt_api_key}&fq={filter_query}&sort={sort}&fl={field_list}&begin_date={begin_date}&end_date={end_date}"
```

```python
# Create an empty list to store the reviews
reviews_list = []

# loop through pages 0-19
for page in range(0,20):
    # create query with a page number
    # API results show 10 articles at a time
    query_url = f"{url}api-key={nyt_api_key}&fq={filter_query}&sort={sort}&fl={field_list}&begin_date={begin_date}&end_date={end_date}"
    query_url = f"{query_url}&page={str(page)}"
    # Make a "GET" request and retrieve the JSON    
    reviews = requests.get(query_url).json()
    
    # Add a twelve second interval between queries to stay within API query limits
    time.sleep(12)
    # Try and save the reviews to the reviews_list
        # loop through the reviews["response"]["docs"] and append each review to the list
    
    try:
        for review in reviews["response"]["docs"]:
            reviews_list.append(review)
        # Print the page that was just retrieved
        print(f"Checked page {page}")

        # Print the page number that had no results then break from the loop
    except KeyError:
        print(f"Nothing found on {page}")
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
data = reviews_list[0:5]
print(json.dumps(data, indent=4))
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()
reviews_list_df = pd.json_normalize(reviews_list)
reviews_list_df.head()
```

```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early
reviews_list_df["title"] = reviews_list_df["headline.main"].apply(lambda st: st[st.find("\u2018")+1:st.find("\u2019 Review")])
reviews_list_df.head()
```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords

# Fix the "keywords" column by converting cells from a list to a string
reviews_list_df['keywords'] = reviews_list_df['keywords'].apply(extract_keywords)
reviews_list_df.head()
```

```python
# Create a list from the "title" column using to_list()
# These titles will be used in the query for The Movie Database
titles = reviews_list_df["title"].to_list()
titles
```

### Access The Movie Database API

```python
# Prepare The Movie Database query
url = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key
search_title = ''
full_details = "https://api.themoviedb.org/3/movie/"
full_api = "?api_key=" + tmdb_api_key

```

```python
# Create an empty list to store the results
tmdb_movies_list = []
# Create a request counter to sleep the requests after a multiple
# of 50 requests
request_counter = 1

# Loop through the titles
for title in titles:
    # Check if we need to sleep before making a request
    if request_counter % 50 == 0:
        print(f"Sleeping at {request_counter} requests")
        time.sleep(1)
        request_counter += 1
    # Add 1 to the request counter
    else:
        request_counter += 1
        search_title = title
        
    # Perform a "GET" request for The Movie Database
    query_url = f"{url}{search_title}{tmdb_key_string}"
    response = requests.get(query_url).json()
    
    # Include a try clause to search for the full movie details.
    # Use the except clause to print out a statement if a movie
    # is not found.
        # try:
        # Get movie id
    try:
        movie_id = response["results"][0]["id"]

        # Make a request for a the full movie details
        full_detail_url = f"{full_details}{movie_id}{full_api}"

        # Execute "GET" request with url
        full_movie_results = requests.get(full_detail_url).json()
        
        # Extract the genre names into a list
        genres = [genre["name"] for genre in full_movie_results.get('genres', [])]

        # Extract the spoken_languages' English name into a list
        spoken_languages = [lang['english_name'] for lang in full_movie_results.get('spoken_languages', [])]

        # Extract the production_countries' name into a list
        production_countries = [country['name'] for country in full_movie_results.get('production_countries', [])]

        # Add the relevant data to a dictionary and
        # append it to the tmdb_movies_list list
        relevant_data = {
            'title' : full_movie_results.get('title'),
            'original_title' : full_movie_results.get('original_title'),
            'budget' : full_movie_results.get('budget'),
            'genres' : genres,
            'original_language' : full_movie_results.get('original_language'),
            'spoken_languages' : spoken_languages,
            'homepage' : full_movie_results.get('homepage'),
            'overview' : full_movie_results.get('overview'),
            'popularity' : full_movie_results.get('popularity'),
            'runtime' : full_movie_results.get('runtime'),
            'revenue' : full_movie_results.get('revenue'),
            'release_date' : full_movie_results.get('release_date'),
            'vote_average' : full_movie_results.get('vote_average'),
            'vote_count' : full_movie_results.get('vote_count'),
            'production_countries' : production_countries
        }
        tmdb_movies_list.append(relevant_data)
        # Print out the title that was found
        print(f"Found {title}")
    except Exception:
        # Handle movie not found
        print(f"{title} not found")
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
tmdb_preview = json.dumps(tmdb_movies_list, indent=4)
print(tmdb_preview)
```

```python
# Convert the results to a DataFrame
tmdb_df = pd.DataFrame(tmdb_movies_list)
tmdb_df
```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews and TMDB DataFrames on title
merge_df = tmdb_df.merge(reviews_list_df, on="title")
merge_df
```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing
columns_to_fix = ["genres","spoken_languages","production_countries"]

# Create a list of characters to remove
remove_characters = ["[","]","'"]

# Loop through the list of columns to fix
for column in columns_to_fix:
    # Convert the column to type 'str'
    merge_df[column] = merge_df[column].astype(str) 

    # Loop through characters to remove
    for char in remove_characters:
        merge_df[column] = merge_df[column].str.replace(char, "")

# Display the fixed DataFrame
merge_df.head()
```

```python
merge_df.columns
```

```python
# Drop "byline.person" column
merge_df = merge_df.drop(columns=['byline.person'], axis=1)
merge_df.columns
```

```python
# Delete duplicate rows and reset index
merge_df = merge_df.drop_duplicates().reset_index(drop=True)
merge_df.head()
```

```python
# Export data to CSV without the index
merge_df.to_csv('movie_review.csv', index=False)
```
